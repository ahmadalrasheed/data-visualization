First of all we have to study the number of people who are smoking from the previous years and how their numbers are changing so that we can have a clear vision about this problem.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('tobacco.csv')
df
```

```python
#the first 23 coloums which represent time of 23 years for one location
new_df=df.head(23)
new_df
```

```python
# draw the relation between years and smoking percentage
plt.scatter(new_df['Year'],new_df['Smoke everyday'],color='red')
plt.xlabel('Years')
plt.ylabel('percentage of smokers')
plt.title('Smoking percentages in 23years')
```

```python
sns.set()

smokers_df=new_df[['Year','Smoke everyday']]
smokers_df
sns.relplot(x="Year",y="Smoke everyday",data=new_df)

```

So this is good indication actually , we recogbnized that during 23 years in one location smoking percentage for people who smokes evry day are getting down on down and this is perfect ! 

Let's see the relation between years and non-smokers percentage.

```python
# draw the relation between years and nonsmokers percentage
sns.relplot(x="Year",y="Never smoked",  data=new_df,color='green')

plt.xlabel('Years')
plt.ylabel('percentage of Non-smokers')
plt.title('Non-Smoking percentages in 23years')
```

## Conclusion
* Smoker number is decreasing year by year
* non smoker numbers is greater that means people is starting to know the negative impact of smoking
